```python
import numpy as np
import matplotlib.pyplot as plt
n=1
a = float(input("enter a number : "))
X = np.linspace(0,a,100)
ax= np.zeros(100)
A = np.sqrt(2/a)
N = (np.pi)*(n/a)
def f(x):
    return A*(np.sin(N*x))
y = f(X)
plt.plot(X,ax)
plt.plot(X , f(X))
plt.show()
```
